```python
import pandas as pd
import numpy as np
```

```python
Customers = pd.read_csv('customers.csv')
order_item = pd.read_csv('order_items.csv')
order = pd.read_csv("orders.csv") 
Product_detail = pd.read_csv("products.csv")
```

```python
Merge = pd.merge(Customers,order,on = "customer_id" )
```

```python
Merge.head(1)
```

```python
Merge2 = pd.merge(order_item,Product_detail,on="product_id")
```

```python
df = pd.merge(Merge,Merge2,on="order_id") 
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date'])
```

```python
df['signup_Month'] = df["signup_date"].dt.month_name()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['order_month'] = df['order_date'].dt.month_name()
```

```python
df['order_year'] = df['order_date'].dt.year
```
